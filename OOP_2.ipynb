Настройте систему логирования с использованием библиотеки logging. Логи должны записываться в файл и в консоль.

```python
# Пример реализации Варианта 14: учет и управление туристическим агентством

from abc import ABC, abstractmethod
from typing import List

# Исключения
class TourServiceError(Exception):
    """Базовое исключение для туристического агентства."""
    pass

class TourNotFoundError(TourServiceError):
    """Исключение при отсутствии тура."""
    pass

class InvalidTourDataError(TourServiceError):
    """Исключение при некорректных данных тура."""
    pass

# Миксины
class LoggingMixin:
    def log(self, message: str):
        print(f"[LOG] {message}")

class NotificationMixin:
    def notify(self, message: str):
        print(f"[NOTIFICATION] {message}")

# Абстрактный класс Tour
class Tour(ABC):
    def __init__(
        self,
        tour_id: int,
        destination: str,
        duration: int,
        price: float,
        is_available: bool = True
    ):
        self.__tour_id = tour_id
        self.__destination = destination
        self.__duration = duration
        self.__price = price
        self.__is_available = is_available

    @abstractmethod
    def calculate_cost(self) -> float:
        """Вычислить полную стоимость тура."""
        pass

    # Геттеры и сеттеры
    def get_id(self) -> int:
        return self.__tour_id

    def get_destination(self) -> str:
        return self.__destination

    def set_destination(self, destination: str):
        if not destination:
            raise InvalidTourDataError("Направление не может быть пустым.")
        self.__destination = destination

    def get_duration(self) -> int:
        return self.__duration

    def set_duration(self, duration: int):
        if duration <= 0:
            raise InvalidTourDataError("Продолжительность должна быть положительным числом.")
        self.__duration = duration

    def get_price(self) -> float:
        return self.__price

    def set_price(self, price: float):
        if price <= 0:
            raise InvalidTourDataError("Цена должна быть положительным числом.")
        self.__price = price

    def is_available(self) -> bool:
        return self.__is_available

    def set_availability(self, available: bool):
        self.__is_available = available

    def __str__(self):
        return (f"Тур (ID {self.get_id()}): {self.get_destination()}, "
                f"длительность {self.get_duration()} дней, цена {self.get_price()}")

    def __lt__(self, other):
        if not isinstance(other, Tour):
            return NotImplemented
        if self.get_price() == other.get_price():
            return self.get_duration() < other.get_duration()
        return self.get_price() < other.get_price()

    def __gt__(self, other):
        if not isinstance(other, Tour):
            return NotImplemented
        if self.get_price() == other.get_price():
            return self.get_duration() > other.get_duration()
        return self.get_price() > other.get_price()

# Конкретные классы туров
class DomesticTour(Tour):
    def calculate_cost(self) -> float:
        # Стоимость = цена за день * количество дней
        return self.get_price() * self.get_duration()

class InternationalTour(Tour):
    def __init__(
        self,
        tour_id: int,
        destination: str,
        duration: int,
        price: float,
        visa_fee: float,
        is_available: bool = True
    ):
        super().__init__(tour_id, destination, duration, price, is_available)
        self.__visa_fee = visa_fee

    def calculate_cost(self) -> float:
        # Стоимость = цена за день * количество дней + консульский сбор
        return self.get_price() * self.get_duration() + self.__visa_fee

    def __str__(self):
        base = super().__str__()
        return f"{base}, виза: {self.__visa_fee}"

# Сервис управления турами
class TourService(LoggingMixin, NotificationMixin):
    def __init__(self):
        self._tours: List[Tour] = []


    def add_tour(self, tour: Tour):
        if not isinstance(tour, Tour):
            raise InvalidTourDataError("Можно добавлять только объекты Tour.")
        self._tours.append(tour)
        self.log(f"Добавлен тур: {tour}")
        self.notify(f"Новый тур: {tour}")

    def edit_tour(self, tour_id: int, **kwargs):
        tour = self._find_by_id(tour_id)
        if 'destination' in kwargs:
            tour.set_destination(kwargs['destination'])
        if 'duration' in kwargs:
            tour.set_duration(kwargs['duration'])
        if 'price' in kwargs:
            tour.set_price(kwargs['price'])
        if 'is_available' in kwargs:
            tour.set_availability(kwargs['is_available'])
        self.log(f"Изменен тур ID {tour_id}")
        self.notify(f"Тур обновлен: {tour}")

    def delete_tour(self, tour_id: int):
        tour = self._find_by_id(tour_id)
        self._tours.remove(tour)
        self.log(f"Удален тур: {tour}")
        self.notify(f"Тур удален: {tour}")

    def get_all_tours(self) -> List[Tour]:
        return list(self._tours)

    def search_by_destination(self, destination: str) -> List[Tour]:
        return [t for t in self._tours if destination.lower() in t.get_destination().lower()]

    def analyze_data(self):
        total = len(self._tours)
        available = sum(1 for t in self._tours if t.is_available())
        avg_cost = (sum(t.calculate_cost() for t in self._tours) / total
                    if total > 0 else 0)
        return {
            'total_tours': total,
            'available_tours': available,
            'average_cost': avg_cost
        }

    def _find_by_id(self, tour_id: int) -> Tour:
        for t in self._tours:
            if t.get_id() == tour_id:
                return t
        raise TourNotFoundError(f"Тур с ID {tour_id} не найден.")

# Демонстрация работы
if __name__ == "__main__":
    service = TourService()

    # Создаем и добавляем туры
    t1 = DomesticTour(1, "Москва", 3, 150.0)
    t2 = InternationalTour(2, "Париж", 5, 200.0, visa_fee=80.0)

    service.add_tour(t1)
    service.add_tour(t2)

    # Редактируем тур
    service.edit_tour(1, price=180.0)

    # Поиск по направлению
    results = service.search_by_destination("пар")
    print("\nПоиск по 'пар':")
    for t in results:
        print(t, "- Стоимость:", t.calculate_cost())

    # Анализ данных
    stats = service.analyze_data()
    print("\nСтатистика по турам:", stats)
```
