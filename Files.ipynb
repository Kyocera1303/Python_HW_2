```python
# Задача 1: Оставить в слове только первые вхождения каждой буквы
def unique_letters(word):
    seen = set()
    result = []
    for char in word:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return ''.join(result)

# Задача 2: Найти длину наибольшей последовательности цифр
def max_digit_sequence(s):
    max_len = current = 0
    for char in s:
        if char.isdigit():
            current += 1
            max_len = max(max_len, current)
        else:
            current = 0
    return max_len

# Задача 3: Проверка двоичного числа и перевод в десятичную систему
def binary_to_decimal():
    while True:
        binary = input("Введите двоичное число: ")
        if all(c in '01' for c in binary):
            return int(binary, 2)
        print("Ошибка! Должны быть только 0 и 1. Попробуйте снова.")

# Выполнение всех задач
if __name__ == "__main__":
    # Задача 1
    words = input("Введите слова через пробел: ").split()
    print("Результат задачи 1:")
    for word in words:
        print(unique_letters(word))

    # Задача 2
    s = input("\nВведите строку с цифрами и буквами: ")
    print(f"Длина самой длинной цифровой последовательности: {max_digit_sequence(s)}")

    # Задача 3
    print("\nРезультат задачи 3:", binary_to_decimal())
```

```python
# Решение всех задач варианта 14

def task1():
    """Задача 1: Разность первой и последней компонент файла"""
    import random

    # Генерация и запись чисел в файл
    N = int(input("\nЗадача 1\nВведите количество чисел N: "))
    numbers = [random.uniform(1, 100) for _ in range(N)]

    with open('numbers.txt', 'w') as f:
        f.write('\n'.join(map(str, numbers)))

    # Чтение и вычисление разности
    with open('numbers.txt', 'r') as f:
        lines = f.read().splitlines()
        if len(lines) >= 1:
            first = float(lines[0])
            last = float(lines[-1]) if len(lines) > 1 else first
            print(f"Разность первой и последней компонент: {first - last:.2f}")
        else:
            print("Файл пуст")

def task2():
    """Задача 2: Поиск игрушек для детей 4 и 10 лет"""
    # Создание тестового файла Assort
    toy_data = [
        "Мяч 500 2 5",
        "Кукла 1200 3 8",
        "Конструктор 2500 5 12",
        "Пазл 800 4 10",
        "Машинка 600 1 4",
        "Настольная игра 1500 6 12"
    ]

    with open('Assort.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(toy_data))

    # Поиск подходящих игрушек
    suitable = []
    with open('Assort.txt', 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 4:
                name = parts[0]
                min_age = int(parts[2])
                max_age = int(parts[3])
                if min_age <= 4 and max_age >= 10:
                    suitable.append(name)

    print("\nЗадача 2")
    print("Игрушки для детей 4-10 лет:", ', '.join(suitable) if suitable else "нет таких игрушек")

def task3():
    """Задача 3: Подсчет слов на заданную букву"""
    # Создание файла с текстом
    text = """В лесу родилась ёлочка,
    В лесу она росла.
    Зимой и летом стройная,
    Зелёная была."""

    with open('text.txt', 'w', encoding='utf-8') as f:
        f.write(text)

    # Подсчет слов
    letter = input("\nЗадача 3\nВведите букву для поиска: ").lower()
    count = 0

    with open('text.txt', 'r', encoding='utf-8') as f:
        for line in f:
            for word in line.strip().split():
                if word.lower().startswith(letter):
                    count += 1

    print(f"Слов на букву '{letter}': {count}")

# Выполнение всех задач
if __name__ == "__main__":
    task1()
    task2()
    task3()
```
