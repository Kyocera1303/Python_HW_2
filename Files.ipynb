```python
#  1:        
def unique_letters(word):
    seen = set()
    result = []
    for char in word:
        if char not in seen:
            seen.add(char)
            result.append(char)
    return ''.join(result)

#  2:     
def max_digit_sequence(s):
    max_len = current = 0
    for char in s:
        if char.isdigit():
            current += 1
            max_len = max(max_len, current)
        else:
            current = 0
    return max_len

#  3:        
def binary_to_decimal():
    while True:
        binary = input("  : ")
        if all(c in '01' for c in binary):
            return int(binary, 2)
        print("!    0  1.  .")

#   
if __name__ == "__main__":
    #  1
    words = input("   : ").split()
    print("  1:")
    for word in words:
        print(unique_letters(word))

    #  2
    s = input("\n     : ")
    print(f"    : {max_digit_sequence(s)}")

    #  3
    print("\n  3:", binary_to_decimal())
```

```python
#     14

def task1():
    """ 1:      """
    import random

    #      
    N = int(input("\n 1\n   N: "))
    numbers = [random.uniform(1, 100) for _ in range(N)]

    with open('numbers.txt', 'w') as f:
        f.write('\n'.join(map(str, numbers)))

    #    
    with open('numbers.txt', 'r') as f:
        lines = f.read().splitlines()
        if len(lines) >= 1:
            first = float(lines[0])
            last = float(lines[-1]) if len(lines) > 1 else first
            print(f"    : {first - last:.2f}")
        else:
            print(" ")

def task2():
    """ 2:     4  10 """
    #    Assort
    toy_data = [
        " 500 2 5",
        " 1200 3 8",
        " 2500 5 12",
        " 800 4 10",
        " 600 1 4",
        "  1500 6 12"
    ]

    with open('Assort.txt', 'w', encoding='utf-8') as f:
        f.write('\n'.join(toy_data))

    #   
    suitable = []
    with open('Assort.txt', 'r', encoding='utf-8') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 4:
                name = parts[0]
                min_age = int(parts[2])
                max_age = int(parts[3])
                if min_age <= 4 and max_age >= 10:
                    suitable.append(name)

    print("\n 2")
    print("   4-10 :", ', '.join(suitable) if suitable else "  ")

def task3():
    """ 3:     """
    #    
    text = """   ,
       .
       ,
     ."""

    with open('text.txt', 'w', encoding='utf-8') as f:
        f.write(text)

    #  
    letter = input("\n 3\n   : ").lower()
    count = 0

    with open('text.txt', 'r', encoding='utf-8') as f:
        for line in f:
            for word in line.strip().split():
                if word.lower().startswith(letter):
                    count += 1

    print(f"   '{letter}': {count}")

#   
if __name__ == "__main__":
    task1()
    task2()
    task3()
```
