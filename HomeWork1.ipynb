```python
# 1.   
def f(x):
    if x <= 13:
        return -x**3 + 9
    else:
        return -3 / (x + 1)

# :
print(f(5))
print(f(20))

# 2.      m, n, p
def count_negatives(m, n, p):
    return sum(1 for number in (m, n, p) if number < 0)

# :
print(count_negatives(-1, 5, -3))

# 3.      
def find_floor(M):
    #  :    3   
    floor = (M + 2) // 3  #  
    if floor % 2 == 0:  #   
        floor += 1
    return floor

# :
print(find_floor(7))

# 4.    t   n
def check_divisibility(t, n):
    if t % n == 0:
        return t // n
    else:
        return f"{t}  {n}   "

# :
print(check_divisibility(10, 2))
print(check_divisibility(7, 3))

# 5. ,     (c, d)
def knight_threatens(a, b, c, d):
    dx = abs(a - c)
    dy = abs(b - d)
    return (dx == 2 and dy == 1) or (dx == 1 and dy == 2)

# :
print(f"  b3: {knight_threatens(1, 1, 2, 3)}")
print(f"  e5: {knight_threatens(4, 4, 5, 5)}")
```

```python
import math

#  1:     
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def gcd_three(a, b, c):
    return gcd(gcd(a, b), c)

#  2:   S = tgx + tgtgx + ... (n )
def calculate_sum(n, x):
    total = 0
    current = x
    for _ in range(n):
        total += math.tan(current)
        current = math.tan(current)
    return total

#  3:     e=0.001
def calculate_series(epsilon=0.001):
    total = 0
    n = 1
    while True:
        numerator = math.factorial(n)
        denominator = math.factorial(3*n + 1)
        term = numerator / denominator
        if term < epsilon:
            break
        total += term
        n += 1
    return total

#  4:   
def is_zigzag(sequence):
    if len(sequence) < 3:
        return False
    for i in range(1, len(sequence)-1):
        prev, curr, next_ = sequence[i-1], sequence[i], sequence[i+1]
        if not (curr > prev and curr > next_) and not (curr < prev and curr < next_):
            return i + 1  # +1       1
    return 0

def process_sequences(K):
    for _ in range(K):
        sequence = []
        while True:
            num = int(input("  (0   ): "))
            if num == 0:
                break
            sequence.append(num)
        result = is_zigzag(sequence)
        if result == 0:
            print(f" ,  : {len(sequence)}")
        else:
            print(f"  ,      : {result}")



choice = input(" : ")

if choice == '1':
            a = int(input("  : "))
            b = int(input("  : "))
            c = int(input("  : "))
            print(f"({a}, {b}, {c}) = {gcd_three(a, b, c)}")

elif choice == '2':
            n = int(input("   (n): "))
            x = float(input("  x ( ): "))
            print(f"S = {calculate_sum(n, x)}")

elif choice == '3':
            print(f" : {calculate_series()}")

elif choice == '4':
            K = int(input("   (K): "))
            process_sequences(K)
```
