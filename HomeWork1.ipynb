```python
# 1. Нахождение значения функции
def f(x):
    if x <= 13:
        return -x**3 + 9
    else:
        return -3 / (x + 1)

# Пример:
print(f(5))
print(f(20))

# 2. Подсчитать количество отрицательных чисел среди m, n, p
def count_negatives(m, n, p):
    return sum(1 for number in (m, n, p) if number < 0)

# Пример:
print(count_negatives(-1, 5, -3))

# 3. На какой этаж должен доставить лифт
def find_floor(M):
    # Номера квартир: этажи идут по 3 квартиры на этаж
    floor = (M + 2) // 3  # Округляем вверх
    if floor % 2 == 0:  # Только нечётные этажи
        floor += 1
    return floor

# Пример:
print(find_floor(7))

# 4. Проверить делимость числа t на число n
def check_divisibility(t, n):
    if t % n == 0:
        return t // n
    else:
        return f"{t} на {n} нацело не делится"

# Пример:
print(check_divisibility(10, 2))
print(check_divisibility(7, 3))

# 5. Определить, угрожает ли конь полю (c, d)
def knight_threatens(a, b, c, d):
    dx = abs(a - c)
    dy = abs(b - d)
    return (dx == 2 and dy == 1) or (dx == 1 and dy == 2)

# Пример:
print(f"Конь угрожает b3: {knight_threatens(1, 1, 2, 3)}")
print(f"Конь угрожает e5: {knight_threatens(4, 4, 5, 5)}")
```

```python
import math

# Задача 1: Наибольший общий делитель трёх чисел
def gcd(a, b):
    while b:
        a, b = b, a % b
    return a

def gcd_three(a, b, c):
    return gcd(gcd(a, b), c)

# Задача 2: Вычисление суммы S = tgx + tgtgx + ... (n слагаемых)
def calculate_sum(n, x):
    total = 0
    current = x
    for _ in range(n):
        total += math.tan(current)
        current = math.tan(current)
    return total

# Задача 3: Сумма ряда с точностью e=0.001
def calculate_series(epsilon=0.001):
    total = 0
    n = 1
    while True:
        numerator = math.factorial(n)
        denominator = math.factorial(3*n + 1)
        term = numerator / denominator
        if term < epsilon:
            break
        total += term
        n += 1
    return total

# Задача 4: Анализ пилообразных последовательностей
def is_zigzag(sequence):
    if len(sequence) < 3:
        return False
    for i in range(1, len(sequence)-1):
        prev, curr, next_ = sequence[i-1], sequence[i], sequence[i+1]
        if not (curr > prev and curr > next_) and not (curr < prev and curr < next_):
            return i + 1  # +1 потому что нумерация элементов начинается с 1
    return 0

def process_sequences(K):
    for _ in range(K):
        sequence = []
        while True:
            num = int(input("Введите число (0 для завершения набора): "))
            if num == 0:
                break
            sequence.append(num)
        result = is_zigzag(sequence)
        if result == 0:
            print(f"Набор пилообразный, количество элементов: {len(sequence)}")
        else:
            print(f"Набор не пилообразный, первый не зубчатый элемент на позиции: {result}")



choice = input("Ваш выбор: ")

if choice == '1':
            a = int(input("Введите первое число: "))
            b = int(input("Введите второе число: "))
            c = int(input("Введите третье число: "))
            print(f"НОД({a}, {b}, {c}) = {gcd_three(a, b, c)}")

elif choice == '2':
            n = int(input("Введите количество слагаемых (n): "))
            x = float(input("Введите значение x (в радианах): "))
            print(f"S = {calculate_sum(n, x)}")

elif choice == '3':
            print(f"Сумма ряда: {calculate_series()}")

elif choice == '4':
            K = int(input("Введите количество наборов (K): "))
            process_sequences(K)
```

```python
import math

# Задача 1: Числа, делящиеся на свои цифры
print("Задача 1: Найти числа, делящиеся на каждую из своих цифр")
n = 15
result_task1 = []
for num in range(1, n + 1):
    digits = [int(d) for d in str(num)]
    if 0 in digits:
        continue
    if all(num % d == 0 for d in digits):
        result_task1.append(num)
print(f"Числа ≤ {n}, делящиеся на все свои цифры: {result_task1}\n")

# Задача 2: Вычитание дробей
print("Задача 2: Вычитание дробей с получением несократимой дроби")
A = int(input("Введите числитель первой дроби (A): "))
B = int(input("Введите знаменатель первой дроби (B): "))
C = int(input("Введите числитель второй дроби (C): "))
D = int(input("Введите знаменатель второй дроби (D): "))

numerator = A * D - B * C
denominator = B * D
common_divisor = math.gcd(numerator, denominator)
simplified_num = numerator // common_divisor
simplified_den = denominator // common_divisor
print(f"Результат вычитания: {simplified_num}/{simplified_den}\n")

# Задача 3: Дни обучения Незнайки
print("Задача 3: Расчет дней обучения Незнайки")
days = 2
words_prev_prev = 1  # первый день
words_prev = 1       # второй день

while True:
    current_words = (words_prev_prev + words_prev) / 2 + 2
    days += 1
    if current_words >= 50:
        break
    words_prev_prev, words_prev = words_prev, current_words

print(f"Силы Незнайки иссякнут через {days} дней")
```
